# [Problem01 ] Implementation using arithmetic operators of powers


```python
%%timeit
# [Problem01 ]Implementation using arithmetic operators of powers

#[Problem 4] Comparison of calculation times
import time
start = time.time()

THICKNESS = 0.00008 #m
DISTANCE_TO_MOON_KM = 384400 #Km

folded_thickness = THICKNESS * (2**43)

#<<<< [Question2] Unit Conversion >>>>
#Convert the thickness to km 
folded_thickness_km = folded_thickness / 1000

print("Thickness: {}m meters".format(folded_thickness))

print("Thickness: {:.2f} kilometers".format(folded_thickness_km))

if(folded_thickness_km >= DISTANCE_TO_MOON_KM):
    print("The folded paper is thicker than the distance to the Moon!")
else:
    print("The folded paper is not thick enough to reach the Moon.")
    
elapsed_time =  time.time() - start
print("time : {}[s])".format(elapsed_time))
```

# [Question 3] Created using a for statement 


```python
#%%timeit
#<<<< [Question 6] Displaying a line graph   >>>>  

 #<<<< [Question 3] Created using a for statement  >>>>
    #[Problem 4] Comparison of calculation times
import time
start = time.time()

THICKNESS = 0.00008 #m
DISTANCE_TO_MOON_KM = 384400 #Km

thickness= THICKNESS
folded_thicknesses = [0]*43

for i in range(43): 
    thickness*=2
    folded_thicknesses[i]=thickness #[Problem 5] Saving to a list
    
print(len(folded_thicknesses)) 
folded_thickness = folded_thicknesses[-1]

import matplotlib.pyplot as plt
%matplotlib inline

#
plt.title("tickness of folded paper")
plt.xlabel("numbeer of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thicknesses) # put the variable name of the list in “list 
plt.show()
#<<<< [Question2] Unit Conversion >>>>
#Convert the thickness to km 
folded_thickness_km = folded_thickness / 1000

print("Thickness: {}m meters".format(folded_thickness))

print("Thickness: {:.2f} kilometers".format(folded_thickness_km))

if(folded_thickness_km >= DISTANCE_TO_MOON_KM):
    print("The folded paper is thicker than the distance to the Moon!")
else:
    print("The folded paper is not thick enough to reach the Moon.")
    
#elapsed_time =  time.time() - start
#print("time : {}[s])".format(elapsed_time))
```
